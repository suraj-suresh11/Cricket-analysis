```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
import requests
```

## Preparing data frame which has batting record data

```python
url = 'https://www.espncricinfo.com/records/most-runs-in-career-282827'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
soup.find('table', class_ = 'ds-w-full ds-table ds-table-xs ds-table-auto ds-w-full ds-overflow-scroll ds-scrollbar-hide')
```

```python
Column_names = table.find_all('span')
print(Column_names)
```

```python
Column_titles = [title.text.strip() for title in Column_names][:15]
print(Column_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = Column_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
df
```

## Preparing data frame with bowling record data

```python
url = 'https://www.espncricinfo.com/records/most-wickets-in-career-283194'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
table1 = soup.find_all('table')[0]
print(table1)
```

```python
Column_names1 = table1.find_all('span')
print(Column_names1)
```

```python
Column_title = [title.text.strip() for title in Column_names1][:15]
print(Column_title)
```

```python
Column_data = [title.text.strip() for title in Column_names1][16:]
print(Column_data)
```

```python
import pandas as pd
df1 = pd.DataFrame(columns = Column_title)
df1
```

```python
column_data1 = table1.find_all('tr')
```

```python
print(column_data1)
```

```python
for row in column_data1[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df1)
    df1.loc[length] = individual_row_data
df1
```

## Formatting the table

```python
df.rename(columns = {'Player Name': 'Player',
                     'Mat': 'Matches',
                     'Inns': 'Innings',
                     'NO':'Not Outs',
                     'HS':'Highest Score',
                     'Ave': 'Average',
                     'BF':'Balls Faced',
                     'SR':'Strike Rate'
                     },
            inplace = True)
df
```

```python
df.replace('-', '0', inplace=True)
df.head(10)
```

```python
df1.rename(columns = {'Mat':'Matches',
                     'Inns':'Innings',
                     'Mdns': "Maidens",
                     'Wkts':'Wickets',
                     'BBI':'Best Bowling Figures',
                     'Ave':'Bowling AVG',
                     '5':'5 Wickets',
                     'Econ':'Economy',
                     '4':'4 Wickets',
                     'SR': 'Strike Rate'
                    },
            inplace = True)
df1
```

```python
df1.replace('-', '0', inplace=True)
df1.head(10)
```
